```python
from pyspark.sql import Row

# Sample data
data = [
    Row(sales_id=1, product="Laptop", amount=1200.00, date="2023-10-01"),
    Row(sales_id=2, product="Phone", amount=800.00, date="2023-10-02"),
    Row(sales_id=3, product="Tablet", amount=400.00, date="2023-10-03")
]

# Create DataFrame
sales_df = spark.createDataFrame(data)
sales_df.display()
```

```python
delta_table_path = "/tmp/delta/sales/"

# Write the DataFrame as a Delta table
sales_df.write.format("delta").mode("overwrite").save(delta_table_path)
#to_test_dep
#12
```
